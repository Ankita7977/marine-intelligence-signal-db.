# Marine Intelligence Signal Normalization

This notebook demonstrates how AIS, weather, and hydrology datasets
are normalized into a unified schema.

**Import Libraries**

```python
import pandas as pd
```

**Load CSV Files**

```python
import pandas as pd

ais = pd.read_csv("../data/ais_standardized.csv")
weather = pd.read_csv("../data/weather_standardized.csv")
water = pd.read_csv("../data/water_standardized.csv")

print("AIS rows:", len(ais))
print("Weather rows:", len(weather))
print("Water rows:", len(water))
```

**Preview Dataset**

```python
ais.head()
```

```python
weather.head()
```

```python
water.head()
```

**Define Unified Schema**

```python
columns = [
"source_id",
"timestamp",
"latitude",
"longitude",
"feature_type",
"normalized_value",
"source_reference"
]
```

**Normalize Datasets**

```python
ais = ais[columns]
weather = weather[columns]
water = water[columns]
```

**Merge Signals**

```python
signals = pd.concat([ais, weather, water], ignore_index=True)

signals.head()
```

**Show Dataset Size**

```python
signals.shape
```

**Feature Types Check**

```python
signals["feature_type"].unique()
```

## Conclusion

The AIS vessel signals, weather precipitation signals, and hydrology
water level signals were successfully normalized into a unified schema.

The normalized dataset contains signals from multiple sources and can
now be ingested into the Marine Intelligence Signal Database for
spatial and temporal analysis.
